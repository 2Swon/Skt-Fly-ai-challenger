```python
class SimplePathFindingEnv:
    def __init__(self, grid):
        self.grid = grid  # 격자
        self.n_rows = len(grid)
        self.n_cols = len(grid[0])
        self.n_states = self.n_rows * self.n_cols
        self.n_actions = 4  # up, down, left, right
        self.current_state = None
        self.goal_state = self._find_goal_state()

    def reset(self):
        while True:
            r, c = random.randint(0, self.n_rows - 1), random.randint(0, self.n_cols - 1)
            if self.grid[r][c] != 0 and self.grid[r][c] != 12:
                self.current_state = (r, c)
                break
        return self.current_state

    def step(self, action):
      r, c = self.current_state
      if action == 0:  # up
          r = max(r - 1, 0)
      elif action == 1:  # down
          r = min(r + 1, self.n_rows - 1)
      elif action == 2:  # left
          c = max(c - 1, 0)
      elif action == 3:  # right
          c = min(c + 1, self.n_cols - 1)

      # 벽 체크
      if self.grid[r][c] == 0:
          return self.current_state, -1, False  # 현재 상태 유지, 벌점 부여

      # 도착 지점 체크
      done = self.grid[r][c] == 12
      reward = 1 if done else -0.1  # 도착 지점에 도달하면 보상, 그렇지 않으면 약간의 벌점

      self.current_state = (r, c)
      return self.current_state, reward, done

    def render(self):
        for r in range(self.n_rows):
            for c in range(self.n_cols):
                if (r, c) == self.current_state:
                    print('P', end=' ')
                elif self.grid[r][c] == 0:
                    print('W', end=' ')  # 벽
                elif self.grid[r][c] == 12:
                    print('F', end=' ')  # 도착 지점
                else:
                    print('-', end=' ')
            print()

    def _find_goal_state(self):
        for r in range(self.n_rows):
            for c in range(self.n_cols):
                if self.grid[r][c] == 12:
                    return (r, c)

# 환경 인스턴스 생성 및 테스트
env = SimplePathFindingEnv(a)
state = env.reset()
env.render()
```

```python
# 환경 인스턴스 생성
env = SimplePathFindingEnv(a)

# 에피소드 시작
state = env.reset()
env.render()

while True:
    action = random.randint(0, 3)  # 랜덤 액션 선택
    next_state, reward, done = env.step(action)
    print(f"Action: {action}, Reward: {reward}, New State: {next_state}")
    env.render()

    if done:
        print("도착 지점에 도달했습니다!")
        break
```

```python
def policy_eval(P, R, pi, maxiter=30):
    V = np.zeros((maxiter, 12), dtype='float32')

    for i in range(maxiter-1):
        V[i+1] = np.squeeze(
            np.matmul(
                np.expand_dims( pi, 1 ),
                np.expand_dims( R + 0.6 * np.dot(P, V[i]), 2 )))

    return V[maxiter-1]

def policy_upd(P, R, v):
    print(np.squeeze(np.expand_dims( R + 0.6 * np.dot(P, v), 2 )))
    a_idx = np.argmax(np.squeeze(np.expand_dims( R + 0.6 * np.dot(P, v), 2 )), axis=1)
    pi = np.zeros((12, 4), dtype='float32')
    pi[range(12), a_idx] = 1.
    return pi

pi_old = None
pi = np.ones((12, 4), dtype='float32') * 0.25

while not np.all(np.equal(pi_old, pi)):
    pi_old = pi.copy()
    v = policy_eval(P, R, pi)
    pi = policy_upd(P, R, v)

print(pi)
```
